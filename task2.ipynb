```python
male_passenger = df.loc[df['Sex'] == 'male', 'Sex']
male_passenger.count()
```

```python
male_survived_passenger = df.loc[(df['Sex'] == 'male') & (df['Survived'] == 1), 'Sex']
male_survived_passenger.count()
```

```python
male_N_survived_passenger = df.loc[(df['Sex'] == 'male') & (df['Survived'] == 0), 'Sex']
male_N_survived_passenger.count()
```

```python
female_passenger = df.loc[df['Sex'] == 'female', 'Sex']
female_passenger.count()
```

```python
female_survived_passenger = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1), 'Sex']
female_survived_passenger.count()
```

```python
female_N_survived_passenger = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 0), 'Sex']
female_N_survived_passenger.count()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(['Male', 'Female'], [male_passenger.count() ,female_passenger.count()])
plt.title('Count of Gender by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
survival_data = {
    'Gender': ['Male', 'Male', 'Female', 'Female'],
    'Survival': ['Survived', 'Non-Survived', 'Survived', 'Non-Survived'],
    'Count': [
        male_survived_passenger.count(),
        male_N_survived_passenger.count(),
        female_survived_passenger.count(),
        female_N_survived_passenger.count()
    ]
}
```

```python
survival_df = pd.DataFrame(survival_data)
display(survival_df)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Count', hue='Survival', data=survival_df, palette='viridis')
plt.title('Distribution of Survivors by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Survival Status')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df['Status'] = df.apply(
    lambda r: 'Children' if (pd.notnull(r['Age']) and r['Age'] < 16)
    else ('Women' if r['Sex'] == 'female' else 'Men'),
    axis=1
)
table = df.groupby(['Status', 'Pclass']).agg(
    Lost = ('Survived', lambda x: (x == 0).sum()),
    Survived = ('Survived', lambda x: (x == 1).sum()),
    Total_on_Board = ('Survived', 'count')
).reset_index()


table['Class'] = table['Pclass'].map({1: "First", 2: "Second", 3: "Third"})
table = table[['Status', 'Class', 'Lost', 'Survived', 'Total_on_Board']]


display("\nFinal Titanic Survival Table:\n")
display(table.to_string(index=False))
```

```python
table['Group'] = table['Status'] + " - " + table['Class']

x = np.arange(len(table['Group']))
width = 0.35

plt.figure(figsize=(14,6))

plt.bar(x - width/2, table['Lost'], width, label='Lost')

plt.bar(x + width/2, table['Survived'], width, label='Survived')

plt.xticks(x, table['Group'], rotation=45, ha='right')
plt.ylabel("Count of Passengers")
plt.title("Titanic Survival by Status and Class")
plt.legend()
plt.tight_layout()
plt.show()
```
